# Background and Introduction

Biological questions require integrative methods more often than ever. However, gaining enough expertise in each of the methods necessary for a project can become a challenge. Collaborative groups begin to address this problem, but learning enough of a discipline to analyze the data is a necessity. Understanding a Transmission Electron Microscopy (TEM) image can be difficult intially, but an annotation tool could act as a bridge to aid [a new person into understanding enough to analyze the data].

We present Bio-STEMGPT, a chatbot designed to aid the user in recognizing features in a TEM image and explore the segmented data. 

- Problem:
    - 
- Making a TEM chatbot to help analyze biological data. Ideally a scientist would give the chatbot some TEM images and the chatbot could (a) tell the user what the most likely tissue type is based on ultrastructure features and (b) give some stats on those ultrastructure features (# nuclei, avg size, avg perimeter of cell membranes). I'm a very visual and comparative

- Proposed solution:
    - chatbot/interactive interface
    - uses cellpose's segmentation algorithm

- Objective:
    - Make a chatbot that can distinguish nuclei in a TEM image based on quantitative measurements (i.e. density of nuclei + shape of cells) and give some basic info/data viz about them.
    - (Stretch): Just adding more ultrastructure features to be able to distinguish other tissue types
